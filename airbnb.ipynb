```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#analysis for price
price_distribution = sns.histplot(df['price'], kde=False)
plt.title('Distribution of Airbnb Listing Prices')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()
print("it shows the number of listings that fall within certain price ranges. the number of listings in each price range.\n Airbnb listings fall between $100 and $500 per night \n most common price range is 100-200 \n the prices below 200 is mpre than the above 200")
```

```python
# Availability analysis
availability_distribution = sns.histplot(df['availability_365'], kde=False)
plt.title('Distribution of Airbnb Listing Availability')
plt.xlabel('Availability (in days)')
plt.ylabel('Frequency')
avg_median=df['availability_365'].median()
plt.vlines(avg_median, ymin=0, ymax=85, color='orange', linewidth=3, label='Median')
plt.text(s=f'median={avg_median:.2f}', x=40, y=50, color='k', fontsize=14)
plt.show()
print("he distribution of Airbnb listing availability according to availability in days. The average availability is around 100 days, while the median availability is around 50 days.\n short-term stays, with a significant number falling under the 30-day \n The median availability is around 244 days, which is significantly higher than the average availability of around 100 days. ")
```

```python
# Plot latitude and longitude on a scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(df['longitude'], df['latitude'], color='blue', alpha=0.5)
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('Geographical Distribution of Airbnb Listings')
plt.grid(True)
plt.show()
print("he color intensity of each data point indicates the density of listings in that area. Darker blue areas represent a higher concentration of Airbnb listings.\n ")
```

```python

# Define the column(s) for outlier detection
columns_for_outlier_detection = ['price']

# Visualize the distribution of the column(s) using box plots
plt.figure(figsize=(8, 6))
sns.boxplot(data=df[columns_for_outlier_detection])
plt.title('Boxplot of Price')
plt.xlabel('Price')
plt.show()
print("The IQR is larger than the median, indicating a skewed distribution \n The box is slightly longer towards the higher end, which means there might be more expensive listings than affordable ones. \n The data points beyond the whiskers are potential outliers \n the ouutliers could be because data entry errors or unusual proprties \n There are outliers on both the lower and upper ends of the price distribution \n")

# Calculate the Interquartile Range (IQR) for each column
Q1 = df[columns_for_outlier_detection].quantile(0.25)
Q3 = df[columns_for_outlier_detection].quantile(0.75)
IQR = Q3 - Q1

# Define the threshold for outliers
outlier_threshold = 1.5

# Identify outliers using the IQR method
outliers = ((df[columns_for_outlier_detection] < (Q1 - outlier_threshold * IQR)) |
            (df[columns_for_outlier_detection] > (Q3 + outlier_threshold * IQR))).any(axis=1)

# Count the number of outliers
num_outliers = outliers.sum()
print(f'Number of outliers: {num_outliers}')

# Remove outliers from the dataset
cleaned_df = df[~outliers]

# Visualize the distribution of the column(s) after removing outliers
plt.figure(figsize=(8, 6))
sns.boxplot(data=cleaned_df[columns_for_outlier_detection])
plt.title('Boxplot of Price (After Removing Outliers)')
plt.xlabel('Price')
plt.show()
print("he boxplot shows the distribution of price after outliers were removed \n interquartile range is still significant, indicating that there is still a variety of prices available in the data set after removing outliers.\n less skewed  ")
```

```python
# Analyze the correlation matrix to identify relationships between variables
plt.figure(figsize=(10, 8))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix')
plt.show()
print(" the colour represents the strength of the correlation \n darker color indicates a stronger correlation \n for price : This means that there is a slight tendency for listings with more features or better locations to be more expensive. \n Availability: Availability has a weak negative correlation with most other variables.")
```

```python
# Compare the distribution of listings across different neighborhoods
plt.figure(figsize=(12, 6))
sns.countplot(x='neighbourhood', data=df, order=df['neighbourhood'].value_counts().index)
plt.title('Distribution of Listings Across Neighborhoods')
plt.xlabel('Neighborhood')
plt.ylabel('Count of Listings')
plt.xticks(rotation=45, ha='right')
plt.show()
print("he most popular neighborhoods for Airbnb listings are Sixth Ward, Second Ward, Tenth Ward, Fourteenth Ward, Ninth Ward and Eighth Ward \n There are also a number of neighborhoods with far fewer listings, such as the Twelfth Ward and the Fifteenth Ward this could be zone restrictions , lack of amenities, or distance from tourist attractions \n")
```

```python
# Explore seasonal trends in demand
df['month'] = pd.to_datetime(df['last_review']).dt.month
plt.figure(figsize=(12, 6))
sns.boxplot(x='month', y='availability_365', data=df)
plt.title('Seasonal Trends in Availability')
plt.xlabel('Month')
plt.ylabel('Availability (in days)')
plt.show()
print("The months with the highest median availability (around 300 days) are from April to September.\n months with the lowest median availability (around 150 days) are from December to March \n higher demand for Airbnb listings, leading to lower availability.\n There is also a significant variability in availability within each month. \n , indicating that there is a wide range of availability options within each month.")
```
